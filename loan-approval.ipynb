```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(df.drop(columns='Risk_Flag'),
                                                   df['Risk_Flag'],
                                                   test_size=0.25,
                                                   random_state=42)

```

```python
print("Train shape:", X_train.shape, "Test shape:", X_test.shape)
```

```python
cat_features = [col for col in X_train.columns if df[col].dtypes=="O"]
num_features = [col for col in X_train.columns if df[col].dtypes!="O"]
```

```python
num_features, cat_features
```

```python
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.ensemble import RandomForestClassifier
from sklearn.pipeline import Pipeline
```

```python
transformer = ColumnTransformer(
    transformers = [
        ("ohe", OneHotEncoder(drop='first'), cat_features),
        ("num", StandardScaler(), num_features),
    ], remainder = 'passthrough'
)
```

```python
pipeline = Pipeline(
    steps = [
        ("transformers", transformer),
        ("classifier", RandomForestClassifier(random_state=42))
    ]
)
```

```python
pipeline.fit(X_train, y_train)
```

```python
y_pred = pipeline.predict(X_test)
```

```python
import joblib

joblib.dump(pipeline, "lgbm_final.pkl", compress=3)
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
print(f"Accuracy: {accuracy_score(y_pred, y_test)}")
print(classification_report(y_pred, y_test))
```
